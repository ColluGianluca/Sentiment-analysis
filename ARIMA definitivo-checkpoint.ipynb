```python
from pylab import rcParams
rcParams['figure.figsize'] = 18, 8
decomposition = sm.tsa.seasonal_decompose(y, model='additive',period=12)
fig = decomposition.plot()
plt.show()
```

```python
train= differenced_data[:99]
test=differenced_data[99:]
```

```python
import pmdarima as pm

# Carica i dati


# Stima i parametri del modello ARIMA
stepwise_fit = pm.auto_arima(train, start_p=0, start_q=0,
                              test='adf',
                              max_p=20, max_q=20, m=12,
                              d=None, seasonal=True,
                              start_P=0, D=0, trace=True,
                              error_action='ignore',  
                              suppress_warnings=True, 
                              stepwise=True)

# Stampa i parametri stimati
print(stepwise_fit.summary())
```

```python
x = np.arange(test.shape[0])
# plt.scatter(x, test, marker='x')
plt.figure(figsize=(18, 9))
plt.plot(x,test['presenze'], color='g')
plt.plot(x, stepwise_fit.predict(n_periods=test.shape[0]))
plt.title('Actual test samples vs. forecasts')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
predictions=stepwise_fit.predict(n_periods=test.shape[0])
MSE_error = mean_squared_error(test, predictions)
print('Testing Mean Squared Error is {}'.format(MSE_error))
```

## occupacy rate

```python
y['capacitÃ _massima']=8356950
df=y
```

```python
lista=[]
for item in predictions[0]:
    lista.append(item)
print(lista)
```

```python
mesi=[ 'Maggio', 'Giugno', 'Luglio', 'Agosto'
                               , 'Settembre', 'Ottobre', 'Novembre', 'Dicembre']
pres=pd.DataFrame(lista,columns=['presenze'])
print(pres)
pres['mesi']=mesi
occupacy_rate=[]
for i in range (0,8):
    #print(data['presenze'][i])
    x=pres['presenze'][-8:].tolist()
    y2=df['capacitÃ _massima'][-8:].tolist()
    rate=(x[i]/y2[i])
    occupacy_rate.append("{0:.0%}".format(rate))
print(occupacy_rate)
```

```python
occupacy_rate_reale=[]
for i in range (0,8):
    #print(data['presenze'][i])
    x=y['presenze'][-8:].tolist()
    y2=df['capacitÃ _massima'].tolist()
    rate=(x[i]/y2[i])
    occupacy_rate_reale.append("{0:.0%}".format(rate))
print(occupacy_rate_reale)
```
